```python
print("press 1 for list operation\npress 2 for set operation\npress 3 for dictionary operation\npress 4 for tuple operation")
ch1=int(input("enter the choice for for select operation"))
if ch1==1:
    l=[10,20,30,40,'ravi','harshit']
    print("list",l)
    print("press 1 for create list:")
    print("press 2 for update list:")
    print("press 3 for add elements in list:")
    print("press 4 for delete elements from list:")
    ch2=int(input("enter the choice for operation:"))
    if ch2==3:
        print("press 1 for append:")
        print("press 2 for insert:")
        print("press 3 for extend:")
        ch3=int(input("enter the method:"))
        if ch3==1:
            l1=int(input("add elements in the list:"))
            l.append(l1)
            print("list after performing append operation:",l)
        elif ch3==2:
            l2=(int(input("Enter the position ")))
            l3=(eval(input("Enter the element")))
            l.insert(l2,l3)
            print("list after performing insert operation:",l)
        elif ch3==3:
            l4=(eval(input("enter multiple elements:")))
            l.extend(l4)
            print("list after performing extend operation:",l)
        else:
            print("invalid choice")
    elif ch2==4:
        print("press 1 for pop:")
        print("press 2 for remove:")
        print("press 3 for clear:")
        print("press 4 for delete:")
        ch4=int(input("enter a method:"))
        if ch4==1:
            print("press 1 for removing last element:")
            print("press 2 for removing using index:")
            pop1=(int(input("enter a choice ")))
            if pop1==1:
                l.pop()
                print("list after performing pop operartion:",l)
            elif pop1==2:
                l1=int(input("enter index of element you want to remove from the list:"))
                l.pop(l1)
                print("list after performing the pop operation:",l)
            else:
                print("invalid choice")
        elif ch4==2:
            l1=int(input("remove the specific position elements:"))
            l.remove(l1)
            print("list after performing the remove operation:",l)
        elif ch4==3:
            l.clear()
            print("list after performing the clear operation:",l)
        elif ch4==4:
            del l
            print("list is deleted successfully")
        else:
            print("invalid choice")
    elif ch2==1:
        print("create a list:")
        l1=eval(input("enter a elements to create a list:"))
        print("list",l1)
    else:
        print("invalid choice")
if ch1==2:
    s={10,20,30,'ravi','shinde'}
    print("set",s)
    print("press 1 for create set:")
    print("press 2 for add elemets in the set:")
    print("press 3 for delete elements in set:")
    ch2=int(input("enter the choice for operation:"))
    if ch2==1:
        print("create a set:")
        s1=eval(input("enter a elements to create a set:"))
        print("set",s1)
    elif ch2==2:
        print("press 1 for add the elements in the set:")
        print("press 2 for update the elements in the set:")
        ch3=int(input("enter the method"))
        if ch3==1:
            s1=eval(input("enter the values which you want to add in the set:"))
            s.add(s1)
            print("set after the performing the add operation:",s)
        elif ch3==2:
            s2=eval(input("enter the multiple values which you want to add in the set:"))
            s.update(s2)
            print("set after the performing the update operation:",s)
        else:
            print("invalid choice")
    elif ch2==3:
        print("press 1 for pop method:")
        print("press 1 for discard method:")
        print("press 1 for remove method:")
        print("press 1 for delete method:")
        ch3=int(input("enter the choice to select method:"))
        if ch3==1:
            s.pop()
            print("set after performing pop operation:")
        elif ch3==2:
            s1=eval(input("enter the value which you want to remove from set:"))
            s.discard(s1)
            print("set after the performing the discard operation:",s)
        elif ch3==3:
            s2=eval(input("enter the value which you want to remove from set:"))
            s.remove(s2)
            print("set after the performing the discard operation:",s)
        elif ch3==4:
            del s
            print("set is deleted successfully.")
        else:
            ("invalid choice")
    else:
        ("invalid choice")
elif ch1 == 3:
    d = dict(eval(input("Create the dictionary : ")))
    print("Press 1 for Addition operation on dictionary.")
    print("Press 2 for deletion operation on dictionary.")
    print("Press 3 for access the data of dictionary.")
    print("Press 4 to get the information about dictionary.")
    print("\n")
    ch2 = int(input("Enter the choice for operation : "))
    if ch2 == 1:
        print("Press 1 for add values using key.")
        print("Press 2 for update method")
        print("\n")
        ch3 = int(input("Enter a choice to select the addition method : "))
        if ch3 == 1:
            a = eval(input("enter the name for key : "))
            b = eval(input("Enter thr value of key : "))
            d[a]=b
            print("Dictionary after performing operation :",d)
        elif ch3 == 2:
            a = dict(eval(input("Enter the multiple key value pair in dictionary format : ")))
            d.update(a)
            print("Dictionary after performing update operation :",d)
        else:
            print("Sorry...,You enterd wrong choice for method.....")
    elif ch2 == 2:
        print("Press 1 for the delete method.")
        print("Press 2 for the pop method.")
        print("Press 3 for the popitems method.")
        print("Press 4 for the clear method.")
        print("\n")
        ch3 = int(input("Enter the choice for the method : "))
        if ch3 == 1:
            a = eval(input("Enter the key of record which you want to delete from dictionary : "))
            del d[a]
            print("Dictionary after a delete operation",d)
        elif ch3 == 2:
            a = eval(input("Enter the key of record which you want to delete from dictionary : "))
            d.pop(a)
            print("Dictionary after a pop operation",d)
        elif ch3 == 3:
            d.popitem()
            print("Dictionary after a popitem operation",d)
        elif ch3 == 4:
            d.clear()
            print("Dictionary after a clear operation",d," i.e. empty dictionary")
        else:
            print("Sorry...,You enterd wrong choice for method.....")
    elif ch2 == 3:
        print("Press 1 to access all the keys of dictionary.")
        print("Press 2 to access all the values of dictionary.")
        print("Press 3 to access value of perticular key based on key name.")
        print("Press 4 to access all key-value pair of dictionary.")
        print("\n")
        ch3 = int(input("Enter the choice for the method : "))
        if ch3 == 1:
            a = d.keys()
            print("All keys of dictionary are :",a)
        elif ch3 == 2:
            a = d.values()
            print("All values of dictionary are :",a)
        elif ch3 == 3:
            k = eval(input("Enter a key name whose value is to be access : "))
            a = d.get(k)
            print(f"Value of the key {k} is :",a)
        elif ch3 == 4:
            a = d.items()
            print("All key-value pairs of dictionary are :",a)
        else:
            print("Sorry...,You enterd wrong choice for method.....")
    elif ch2 == 4:
        print(" Dictionary is a unordered, heterogenous and mutable group key-value pairs. \n In dictionary keys are immutable while values are mutable. \n Dictionary are enclosed in curly bracket {} and dictionary element are seperated by comma (,)")
    else:
        print("Sorry...,You enterd wrong choice for operation.....")
if ch1 == 4:
    t = eval(input("Create the tuple : "))
    print("Press 1 to access the single elemnt from tuple using indexing.")
    print("Press 2 to access multiple element from tuple using slicing.")
    ch2 = int(input("Enter the choice to select method : "))
    if ch2 == 1:
        i = int(input("Enter the index whose value is to be access : "))
        a = t[i]
        print(f"Value of index {i} is :",a)
    elif ch2 == 2:
        s = int(input("Enter the start index :"))
        e = int(input("Enter the end index : "))
        g = int(input("Enter the step between values : "))
        a = t[s:e:g]
        print("Values after performing slicing according to given indexes :",a)
    else:
        print("invalid choice.")
```
